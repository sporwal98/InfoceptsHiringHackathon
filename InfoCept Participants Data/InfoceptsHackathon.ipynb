```python
lines = open('infocept_xml_data_.xml').readlines()
for i in range(len(lines)):
    lines[i] = lines[i].replace(' ','')
    lines[i] = lines[i].replace('\n','')
lines[:30]
```

```python
lines = lines[2:-1]
```

```python
lists = {'index':[],
         'user_session':[],
         'total_clicks':[],
         'total_items':[],
         'total_cats':[],
    'max_dwell':[],
    'mean_dwell':[],
    'total_duration':[],
    'click_rate':[],
    'day_of_week':[],
    'is_weekend':[],
    'is_special_day':[],
    'time_of_day':[],
    'cat_most_viewed_n_times':[],
    'cat_most_viewed':[],
    'prod_most_viewed_n_times':[],
    'prod_most_viewed':[],
    'cat_views_freqs':[],
    'prod_views_freqs':[],
    'cat_buys_freqs':[],
    'prod_buys_freqs':[],
    'is_purchase':[],
    'start_time_ts':[],
    'end_time_ts':[],
    'row':[],
    '/row':[],
        }
```

```python
for i in range(len(lines)):
    if(np.mod(i-1,1000000)==0):
        print('line',i-1,'/',len(lines),' processed')
    line = lines[i]
    #print(line)
    prop = line[(line.find('<')+1):line.find('>')]
    value = line[(line.find('>')+1):(line.find('</'))]
    #print('prop:',prop,',value:',value)
    lists[prop].append(value)
```

```python
for k in lists.keys():
    print(k,':',len(lists[k]))
```

```python
df = pd.DataFrame(lists)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df.drop(['row','/row'],axis = 1)
```

```python
df.describe()
```

```python
df.columns
```

```python
sub_df = pd.DataFrame()
```

```python
time_spec_points = []
for time in df['start_time_ts']:
    time = time.split(',')
    time = int(time[-1].split(':')[0])
    if(time>=16):
        time+=5
    time_spec_points.append(time)
sub_df['time_spec_points'] = time_spec_points
```

```python
total_duration = np.round(df['total_duration'].apply(lambda x: float(x)),2)
df['total_duration'] = total_duration
```

```python
give_big_discount = np.logical_and(np.logical_not(df.day_of_week),df.is_special_day)
sub_df['give_big_discount'] = give_big_discount
```

```python
sub_df['prod_views_buys_ratio'] = df.prod_views_freqs.apply(lambda x: float(x))/df.prod_buys_freqs.apply(lambda x: float(x))
```

```python
df['total_clicks']=df['total_clicks'].apply(lambda x: int(x))
df['total_items']=df['total_items'].apply(lambda x: int(x))
df['total_cats']  =df['total_cats'].apply(lambda x: int(x))
```

```python
total_inv = df['total_clicks']+df['total_items']+df['total_cats']
```

```python
for i in range(len(total_inv)):
    if(total_inv[i]>10):
        total_inv[i]*=2
```

```python
total_inv.shape
```

```python
sub_df['total_inventory'] = total_inv
```

```python
l_points = []
for i in range(0,len(df)):
    if(np.mod(i-1,10000)==0):
        print('Processed: ',i-1)
    if(total_duration[i]>3 & int(df['day_of_week'][i])==7):
        l_points.append(10*(int(df['day_of_week'])-3))
    else:
        l_points.append(0)
sub_df['loyalty_points'] = l_points
```

```python
submission_sample = pd.read_csv('infocept_submission.csv')
```

```python
sub_cols = submission_sample.columns
```

```python
for col in sub_cols:
    if(col not in sub_df.columns):
        sub_df[col] = df[col]
        
```
